Gráfico de Dispersão - Árvores

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('dados/trees.csv')
base.columns = ["Circunferencia", "Altura", "Volume"]
base.head()
```

```python
plt.scatter(base.Circunferencia, base.Volume, color = 'Blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
plt.plot(base.Circunferencia, base.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
sns.regplot(x = base.Circunferencia, y = base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```

```python
sns.regplot(x = base.Circunferencia, y = base.Volume, data = base, x_jitter = 0.3, fit_reg = True)
```
